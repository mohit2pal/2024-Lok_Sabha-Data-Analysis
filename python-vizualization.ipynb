```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'Data/unified_dataset.csv'
dataset = pd.read_csv(file_path)

# Distribution of total votes
plt.figure(figsize=(10, 6))
sns.histplot(dataset['Total Votes'].dropna(), bins=30, kde=True)
plt.title('Distribution of Total Votes')
plt.xlabel('Total Votes')
plt.ylabel('Frequency')
plt.show()
```

```python

# Distribution of percentage of votes
plt.figure(figsize=(10, 6))
sns.histplot(dataset['% of Votes'].dropna(), bins=30, kde=True)
plt.title('Distribution of Percentage of Votes')
plt.xlabel('Percentage of Votes')
plt.ylabel('Frequency')
plt.show()
```

```python

# Top 10 parties by total votes
top_parties = dataset.groupby('Party')['Total Votes'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_parties.values, y=top_parties.index)
plt.title('Top 10 Parties by Total Votes')
plt.xlabel('Total Votes')
plt.ylabel('Party')
plt.show()
```

```python

# State-wise total votes
state_votes = dataset.groupby('State')['Total Votes'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=state_votes.values, y=state_votes.index)
plt.title('Top 10 States by Total Votes')
plt.xlabel('Total Votes')
plt.ylabel('State')
plt.show()
```

```python
# Calculate total postal votes casted
total_postal_votes = dataset['Postal Votes'].sum()

# Calculate percentage of postal votes with respect to total votes
percentage_postal_votes = (total_postal_votes / dataset['Total Votes'].sum()) * 100

# Print the results
print("Total Postal Votes Casted:", total_postal_votes)
print("Percentage of Postal Votes:", percentage_postal_votes)

# Visualization
plt.figure(figsize=(8, 6))
sns.barplot(x=['Total Votes', 'Postal Votes'], y=[dataset['Total Votes'].sum(), total_postal_votes])
plt.title('Total Votes vs Postal Votes')
plt.xlabel('Vote Type')
plt.ylabel('Vote Count')
plt.show()
```
